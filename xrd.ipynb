Cette tâche consiste à déterminer les 3 premiers pics du diffractogramme correspondant
pour un rayonnement X incident Cu Kα de longueur d’onde λ = 1.54060 Å.
Grâce à la loi de Bragg : 2*d(hkl)sin(theta) = lambda
    

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from tabulate import tabulate

with MPRester("AQwsCqxPeioMztVOij") as m:        
    struct = m.get_structure_by_material_id("mp-961774")
xrd = XRDCalculator(wavelength = 'CuKa',symprec = 2.0)
pattern = xrd.get_pattern(struct)
table = []
for i in range(3):
    pic = pattern.hkls[i][0]
    table.append(["Pic "+str(i), pic["hkl"], pattern.d_hkls[i], pattern.x[i], pattern.y[i]])
    
print(tabulate(table, headers=["Pic","hkl", "d", "Angle de diffraction", "Intensité du pic"]),"\n")    
print(xrd.show_plot(struct))
```
